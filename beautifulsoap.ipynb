```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

# Make the GET request to fetch the webpage
webpage = requests.get('https://www.kashmir-tourism.org/tourist-destinations/srinagar.html', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python

print(soup.prettify())
```

```python


# Find the main section
left_panel = soup.find(id='leftPanel')

# Find all headings (h1) and paragraph-based subheadings (<p><strong>...</strong></p>)
headings = left_panel.find_all(['h1', 'p'])

# Dictionary to store heading and its corresponding paragraphs
content_dict = {}

current_heading = None  # Track current heading

for tag in headings:
    if tag.name == 'h1':  # Main heading
        current_heading = tag.text.strip()
        content_dict[current_heading] = []
    elif tag.name == 'p' and tag.find('strong'):  # Subheadings inside <p><strong>...</strong></p>
        current_heading = tag.text.strip()
        content_dict[current_heading] = []
    elif tag.name == 'p' and current_heading:  # Regular paragraphs under a heading
        content_dict[current_heading].append(tag.text.strip())

# Print extracted data
for heading, paragraphs in content_dict.items():
    print(f"📌 {heading}\n")
    for para in paragraphs:
        print(f"➡ {para}\n")
    print("-" * 50)

```
